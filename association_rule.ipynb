```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd
data=pd.read_csv("Online Retail.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data['Country'].unique()
```

```python
#remove space in desccription strip is used
data['Description']=data['Description'].str.strip()
```

```python
data.isnull().sum()
```

```python
#drop the row that dont have invoice no
data.dropna(axis=0,subset=['InvoiceNo'],inplace=True)
```

```python
#looking at sales for france only for ease
basket=(data[data['Country']=="France"]
        .groupby(['InvoiceNo','Description'])['Quantity'].sum()
        .unstack()
        .reset_index().fillna(0)
        .set_index('InvoiceNo'))
```

```python
basket.shape
```

```python
basket.to_excel(r"C:\Users\YOGA\OneDrive\Desktop\basket.xlsx")
```

```python
#makinga function which returns 0 or 1
def replace_quantity(x):
    if x>=1:
        return 1
    else:
        return 0
```

```python
basket_sets=basket.applymap(replace_quantity)
```

```python
basket_sets.to_excel(r"C:\Users\YOGA\OneDrive\Desktop\basket_data.xlsx")
```

```python
basket_sets.drop('POSTAGE',inplace=True,axis=1)
```

```python
frequent_itemsets=apriori(basket_sets,min_support=0.07,use_colnames=True)
```

```python
rules=association_rules(frequent_itemsets,metric="lift",min_threshold=1)
rules
```

```python
rules[(rules['lift']>=6) & (rules['confidence']>=0.8)]
```
