```python
import seaborn as sns

flights=sns.load_dataset("flights")
flights_pivoted=flights.pivot(index="month",columns="year",values="passengers")
ax=sns.heatmap(flights_pivoted,cmap="YlGnBu")
```

```python
ax=sns.heatmap(flights_pivoted,annot=True,fmt="d")
```

```python
ax=sns.heatmap(flights_pivoted,cmap="crest",linecolor="white",linewidths=3)
```

# clustor map

```python
sns.clustermap(flights_pivoted)
```

```python
sns.clustermap(flights_pivoted,cmap="coolwarm",standard_scale=1)
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
g=sns.FacetGrid(tips,col="sex")
g.map(sns.scatterplot,"total_bill","tip")
```

# pairGrid

```python
penguins = sns.load_dataset("penguins")
g=sns.PairGrid(penguins,diag_sharey=False)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot)
```

```python
sns.jointplot(data=penguins,x="bill_length_mm",y="bill_depth_mm")
```

```python
g=sns.JointGrid(data=penguins,x="bill_length_mm",y="bill_depth_mm")
g.plot(sns.scatterplot,sns.histplot,alpha=.7,edgecolor=".2",linewidth=.5)
```
