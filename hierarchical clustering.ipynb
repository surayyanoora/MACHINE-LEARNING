```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=pd.read_csv("Mall_Customers.csv")
```

```python
data.head()
```

```python
X=data.iloc[:,[3,4]].values
```

```python
from sklearn.metrics.pairwise import euclidean_distances
diatance_matrix = euclidean_distances(X)
distance_df = pd.DataFrame(diatance_matrix)#convert distace matrix to dataframe for better visualization
#set raws and columns for better interpretation
distance_df.columns = [f'Distance to point {i+1}' for i in range(distance_df.shape[1])]
distance_df.index = [f'point {i+1}'  for  i in range(distance_df.shape[0])]
                                                  
distance_df
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title("Dendogram")
plt.xlabel("Customers")
plt.ylabel("Eucledean diatances")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=4,linkage='ward')
```

```python
y_hc=hc.fit_predict(X)
```

```python
y_hc
```

```python

plt.scatter(X[y_hc==0,0], X[y_hc==0,1], s = 100, c ='red', label='cluster 1')
plt.scatter(X[y_hc==1,0], X[y_hc==1,1], s = 100, c ='blue', label='cluster 2')
plt.scatter(X[y_hc==2,0], X[y_hc==2,1], s = 100, c ='green', label='cluster 3')
plt.scatter(X[y_hc==3,0], X[y_hc==3,1], s = 100, c ='cyan', label='cluster 4')

plt.title("Cluster Of Customers")
plt.xlabel("Annual Income (k$))")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show() 
```
