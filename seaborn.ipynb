```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset('car_crashes')
df.head()
```

```python
# styling

plt.scatter(df.speeding,df.alcohol)
plt.show()
```

```python
sns.set_style("white")
plt.subplot(151)
sns.set_style("ticks")
plt.subplot(152)
sns.set_style("dark")
plt.subplot(153)
sns.set_style("whitegrid")
plt.subplot(154)
sns.set_style("darkgrid")
plt.subplot(155)
plt.show()
```

# regression plot


```python
import seaborn as sns
tips = sns.load_dataset("tips")
ax=sns.regplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
ax=sns.regplot(x="total_bill",y="tip",data=tips,x_jitter=0.8)
plt.show()
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
ax=sns.lmplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,col="day",hue="sex",col_wrap=2,height=3)
plt.show()
```

```python
flights=sns.load_dataset("flights")
flights_pivoted=flights.pivot(index="month",columns="year",values="passengers")
ax=sns.heatmap(flights_pivoted,cmap="YlGnBu")
```

```python
ax=sns.heatmap(flights_pivoted,annot=True,fmt="d")
```
