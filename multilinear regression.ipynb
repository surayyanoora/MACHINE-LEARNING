# removing outliers

```python
Q1=data['Price'].quantile(0.25)
Q3=data['Price'].quantile(0.75)
IQR = Q3-Q1
filter_price= (data['Price'] >= Q1-3.5*IQR) & (data['Price'] <= Q3 + 3.5*IQR)
data =data.loc[filter_price]


plt.subplot(1,2,1)
p=sns.boxplot(y='Price',data=data)

plt.subplot(1,2,2)
q=sns.histplot(x='Price',data=data)
plt.tight_layout()
```

# correltion analysis

```python
corr_matrix =  data.drop(columns=['Location']).corr()
plt.figure(figsize=(28,16))
sns.heatmap(data=corr_matrix,annot=True)
plt.show()
```

# feature scaling

```python
data["Area"]=np.log1p(data['Area'])
```

```python
data["Price"]=np.log1p(data['Price'])
```

# selecting higher correlated data

```python
corr_matrix=data.drop(columns=['Location']).corr()

correlation_threshold = 0.34
high_corr_col = corr_matrix.loc[(corr_matrix['Price'] >correlation_threshold),'Price'].index.tolist() 
print(f"columns with correlation aboue {correlation_threshold} with 'price'")
print(high_corr_col)
high_corr_data=data[high_corr_col]
corr_matrix_high_corr = high_corr_data.corr()

plt.figure(figsize=(10,8))
sns.heatmap(data=corr_matrix_high_corr ,annot=True,cmap= 'coolwarm')
plt.title(print(f"correlation heatmap for columns with correlation > {correlation_threshold} with 'price'"))
plt.show()
```

# model training

```python
x=high_corr_data.drop(['Price'],axis='columns').values
y=high_corr_data['Price']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error

model = LinearRegression()

model.fit(X_train,y_train)

y_predt = model.predict(X_test)
```

```python
r2 = r2_score(y_test,y_predt)
print(r2)
```

```python
# adjusted r2
r2 = r2_score(y_test,y_predt)
n=len(y_test)
p=X_train.shape[1]
adjusted_r2=1-(1-r2) * (n-1) / (n-p-1)
print(adjusted_r2)
```

```python
mae= np.mean(np.abs(y_predt-y_test))
mse= mean_squared_error(y_test,y_predt)
rmse=np.sqrt(mse)

print("Mean absolute error",mae)
print("Mean squared error",mse)
print("Root mean sq error",rmse)
```

# saving model
